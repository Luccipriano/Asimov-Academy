## Aula de Variáveis

Uso de variáveis facilita a leitura, desenvolvimento e manutenção do código.

```python
print(3.14 * 5 ** 2)

pi = 3.14
raio_circunferencia = 5
raio_ao_quadrado = raio_circunferencia ** 2

print("Raio da Circcunferência:",pi * raio_ao_quadrado)
```

Variáveis são sensíveis a capitalização.

```python
a = 4
print(A)
```

```python
a = 4
print(a)
```

```python
nome = 'Lucas'
print(Nome)
```

Variáveis não podem começar com número, mas podem conter números

```python
a1 = 50
print(a1)
```

```python
1A = 50
print(1A)
```

Existem certas palavras e termos reservados que não podem ser usados como variáveis.

```python
if = 2
```

Funções (não apenas funções: *built-ins*, identificadores, qualquer nome definido por um escopo maior) podem ser sobrescritas por definições de variáveis, é o chamado variable shadowing.

>Variable shadowing é quando um nome local (ou de escopo mais interno) esconde um nome com o mesmo identificador definido em um escopo externo (builtin, módulo global, ou escopo fechador).

```python
print('Oi!')

print = 'Tudo bem?'

print(5)
```
