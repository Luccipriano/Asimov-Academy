# Dataframe

````python
import numpy as np
import pandas as pd

from numpy.random import randn
````

````python
df = pd.DataFrame(randn(5, 4), index=["A", "B", "C", "D", "E"], columns="W X Y Z".split())
df
````

````python
print(df["W"])
type(df["W"])
````

````python
print(df[["W", "Y"]])
type(df[["W", "Y"]])
````

````python
df["Novo"] = df["W"]
df
````

````python
df["Novo"] = df["X"] + df ["Y"]
df
````

````python
df["Novo"].drop("A", axis=0) ## Função de DROP similar a cláusula do SQL
## Nesse caso ao fazer df["Novo"] eu estou convertendo e selecionando uma series do meu Dataframe, ao dizer drop("A", axis=0), eu estou selecionando para dropar o índice A eixo 0 que é exatamente a 1ª posição da 1ª (e única coluna)

df[["Novo"]].drop("Novo", axis=1)
# Agora eu preservei o objeto Dataframe porque eu selecionei a coluna "Novo" dentro de uma matriz
# Ao dropar "Novo" eu estou selecionando a coluna toda e o eixo 1 é o eixo das colunas

#df[["Novo"]].drop("Novo", axis=0) -> Não funciona porque axis=0 são os índices pra um DF, não existe o índice "Novo" nesse conjunto de dados, apenas A, B, C...

df[["Novo"]].drop("A", axis=0)
# axis=0 seleciona os índices de um DF, logo, eu consigo acessar e dropar o índice A ou B ou C ou D etc
````

````python
df ## O Dataframe original não sofreu mudanças porque o drop foi apenas executado pra gerar output e não foi armazenado em nenhuma variável, incluindo a original
````

````python
df.drop("A", axis=0, inplace=True)
df
# O parâmetro inplace vai manipular se estou alterando no DF original ou não, caso não queira usar uma nova variável
````

````python
df.loc[["B"]] # Consigo selecionar/localizar/extrair partes específicas assim também
````

````python
df.iloc[0,0] # Trabalha como se fosse um localizador via "coordenada"
````

#### Obs.: sobre manipulação de índices em colunas ->

Por mais que não exista um comando "reset_columns", as colunas são índices assim como o que de fato é manipulado no reset_index.

É possível manipular os índices das colunas de duas formas:

##### Cenário 1: O "Truque da Transposição" (Literalmente resetar as colunas)
Se você quer pegar os nomes das suas colunas e transformá-los em uma linha de dados dentro do DataFrame (exatamente o que o reset_index faz com o índice das linhas), a maneira mais criativa e direta é transpor o DataFrame, resetar o índice e transpor de volta.

```python
import pandas as pd

"# Criando um DataFrame simples
df = pd.DataFrame({
    'Matemática': [90, 85],
    'História': [70, 75],
    'Física': [88, 92]
}, index=['Aluno A', 'Aluno B'])

print("--- Original ---")
print(df)

"# O TRUQUE:
"# 1. .T (Transpõe: colunas viram linhas)
"# 2. .reset_index() (O índice - antigas colunas - vira uma coluna de dados)
"# 3. .T (Transpõe de volta)
df_resetado = df.T.reset_index().T

print("\n--- Com Colunas 'Resetadas' para dentro dos dados ---")
print(df_resetado)
```

##### Cenário 2: Achatando Multi-Index (O caso mais comum)
Se você fez um groupby ou uma pivot table, provavelmente acabou com colunas hierárquicas (MultiIndex), onde você tem "níveis" de colunas (ex: ('Vendas', '2023'), ('Vendas', '2024')).

Nesse caso, o "reset" que a maioria das pessoas deseja é achatar esses níveis para ter apenas uma linha de cabeçalho simples.

```python
import pandas as pd

"# Criando um DataFrame com MultiIndex nas colunas (simulando um groupby)
cols = pd.MultiIndex.from_tuples([("Vendas", "Jan"), ("Vendas", "Fev"), ("Lucro", "Jan"), ("Lucro", "Fev")])
data = [[100, 120, 10, 12], [110, 130, 11, 13]]
df_multi = pd.DataFrame(data, columns=cols)

print("--- Original (MultiIndex) ---")
print(df_multi)

"# SOLUÇÃO: Achatar os níveis
"# Isso une os níveis com um sublinhado (ex: Vendas_Jan)
df_multi.columns = ['_'.join(col).strip() for col in df_multi.columns.values]

"# Se quiser apenas remover um nível específico (ex: ficar só com Jan/Fev e ignorar Vendas/Lucro):
"# df_multi.columns = df_multi.columns.get_level_values(1)

print("\n--- Achatado (Nível único) ---")
print(df_multi)
```
