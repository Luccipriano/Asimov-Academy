# Dataframe

```python
import numpy as np
import pandas as pd

from numpy.random import randn
```

```python
df = pd.DataFrame(randn(5, 4), index=["A", "B", "C", "D", "E"], columns="W X Y Z".split())
df
```

```python
print(df["W"])
type(df["W"])
```

```python
print(df[["W", "Y"]])
type(df[["W", "Y"]])
```

```python
df["Novo"] = df["W"]
df
```

```python
df["Novo"] = df["X"] + df ["Y"]
df
```

```python
df["Novo"].drop("A", axis=0) ## Função de DROP similar a cláusula do SQL
## Nesse caso ao fazer df["Novo"] eu estou convertendo e selecionando uma series do meu Dataframe, ao dizer drop("A", axis=0), eu estou selecionando para dropar o índice A eixo 0 que é exatamente a 1ª posição da 1ª (e única coluna)

df[["Novo"]].drop("Novo", axis=1)
# Agora eu preservei o objeto Dataframe porque eu selecionei a coluna "Novo" dentro de uma matriz
# Ao dropar "Novo" eu estou selecionando a coluna toda e o eixo 1 é o eixo das colunas

#df[["Novo"]].drop("Novo", axis=0) -> Não funciona porque axis=0 são os índices pra um DF, não existe o índice "Novo" nesse conjunto de dados, apenas A, B, C...

df[["Novo"]].drop("A", axis=0)
# axis=0 seleciona os índices de um DF, logo, eu consigo acessar e dropar o índice A ou B ou C ou D etc
```

```python
df ## O Dataframe original não sofreu mudanças porque o drop foi apenas executado pra gerar output e não foi armazenado em nenhuma variável, incluindo a original
```

```python
df.drop("A", axis=0, inplace=True)
df
# O parâmetro inplace vai manipular se estou alterando no DF original ou não, caso não queira usar uma nova variável
```

```python
df.loc[["B"]] # Consigo selecionar/localizar/extrair partes específicas assim também
```

```python
df.iloc[0,0] # Trabalha como se fosse um localizador via "coordenada"
```

# ILoc e Filtros

```python
df
df.loc[["E", "D"],["X", "Z"]]
```

```python
df.iloc[:-1] # O ":" denota um invtervalo e o inteiro (1, 2, -1, -2, etc) representa o ponto de início do intervalo, que vai "iterar"/exibir os dados até o fim do data frame
# Valores positivos iniciam do "começo" do DataFrame (1ª linha), valores negativos a partir do "fim" (última linha)
```

```python
df.iloc[:1, :3] # Passar um segundo argumento vai denotar a segunda "coordenada" do Data Frame
```

```python
df.iloc[:2, 1:2] # Seleção de lista do tipo "1:2", o 1 é inclusivo e o 2 é exclusivo. Isto é, o que vem antes do ":" inclui o índice 1 e o que vem depois exclui o índice 2
```

```python
df > 0 # Executa uma operação de comparação matricial com um escalar
```

```python
df[df > 0]
# Eu consigo também fazer seleções que levem em conta comparações lógicas e, além disso, usar dataframes pra selecionar dataframes
# O que tá acontecendo aqui é que o "df > 0" está retornando um dataframe com true e false e o "df[]" está usando isso pra saber que valores selecionar e quais não selecionar
```

```python
df
```

```python
df["X"] > 0 # Posso fazer essa mesma manipulação com Series ao invés do dataframe todo
```

```python
df[df["X"] > 0] # O índice E foi eliminado da seleção, pois na célula acima o E assume valor falso na comparação, o objeto dataframe está recebendo uma lista de true e false associada aos índices que ele precisa filtrar
```
