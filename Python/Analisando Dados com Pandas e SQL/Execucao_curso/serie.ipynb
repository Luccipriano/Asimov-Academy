## Aula de Séries

```python
import pandas as pd
import numpy as np
```

```python
labels = ['a', 'b', 'c'] ## Isso é uma lista

minha_lista = list((10,20,30)) ## Isso também é uma lista, sendo criada por meio da função "list()"

arr = np.array([10,20,30]) ## Isso transforma a lista em um array numpy

d = {'a': 10, 'b': 20, 'c': 30}
```

```python
pd.Series(labels) ## Criando uma series a partir de uma lista padrão do python
```

```python
pd.Series(data=labels, index=minha_lista) ## Posso manipular a criação dos indíces da minha series a partir de seus parâmetros
```

```python
pd.Series(labels, minha_lista) 
## Eu também não preciso dizer os parâmetros se eles estiverem na mesma ordem padrão da assinatura da função.
```

```python
d ## A minha variável D é um dicionário
```

```python
pd.Series(d) ## Quando eu passei um dicionário pra criar uma series ele já inferiu que as 'chaves' do dicionário seriam os índices da series
```
