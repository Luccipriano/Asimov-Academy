# Desafio - crie um programa que:
- Pede pelo seu nome e idade
- Dá oi para você
- Conta quantas letras seu nome possuí
- Fala quantos anos você terá daqui a 5 anos

```python
# Coleta de dados do usuário
nome = input("Diga seu nome:")
idade = input("Diga sua idade em anos:")
idade_como_int = int(idade)

# Processamento: calcula o tamanho do nome e a idade futura, exibindo os resultados
tamanho_nome = len(nome)
print("Oi,", nome+".")
print("Seu nome possui", tamanho_nome, "letras.")
print("Em cinco anos você terá", idade_como_int + 5, "anos.")
```
